## UPDATE BY LOAD_DATE

```python
%%sql
-- Create Delta table for TRANSFORMED_PIPELINE_DASHBOARD
CREATE TABLE IF NOT EXISTS TRANSFORMED_PIPELINE_DASHBOARD (
    ID STRING,
    ID_TASKS STRING,
    DEPARTMENT STRING,
    KPI STRING,
    CATEGORY_BREAKDOWN STRING,
    TASKS STRING,
    MEASUREMENTS STRING,
    TYPE STRING,
    EMPLOYEES STRING,
    VALUE DOUBLE,
    JAN STRING,
    FEB STRING,
    MAR STRING,
    APR STRING,
    MAY STRING,
    JUNE STRING,
    JULY STRING,
    AUGUST STRING,
    SEPTEMBER STRING,
    OCTOBER STRING,
    NOV STRING,
    DEC STRING,
    COMPLETED_TASK LONG,
    TOTAL_TASK LONG,
    LOAD_DATE DATE 
) USING DELTA
PARTITIONED BY (LOAD_DATE);
```

## INSERT BY LOAD_DATE

```python
%%sql
MERGE INTO TRANSFORMED_PIPELINE_DASHBOARD AS target
USING transformed_pipeline AS source
ON target.ID = source.ID
WHEN MATCHED THEN
    UPDATE SET
        target.ID_TASKS = source.ID_TASKS,
        target.TYPE = source.Type,
        target.TASKS = source.Tasks,
        target.VALUE = source.Value,
        target.JAN = source.Jan,
        target.FEB = source.Feb,
        target.MAR = source.Mar,
        target.APR = source.Apr,
        target.MAY = source.May,
        target.JUNE = source.June,
        target.JULY = source.July,
        target.AUGUST = source.August,
        target.SEPTEMBER = source.September,
        target.OCTOBER = source.October,
        target.NOV = source.Nov,
        target.DEC = source.Dec,
        target.COMPLETED_TASK = source.Completed_task,
        target.TOTAL_TASK = source.Total,
        target.DEPARTMENT = source.Department,
        target.EMPLOYEES = source.Employees,
        target.KPI = source.KPI,
        target.CATEGORY_BREAKDOWN = source.Category_Breakdown,
        target.MEASUREMENTS = source.Measurements,
        target.LOAD_DATE = source.LOAD_DATE

WHEN NOT MATCHED THEN
    INSERT (
        ID_TASKS,
        TYPE,
        TASKS,
        VALUE,
        JAN,
        FEB,
        MAR,
        APR,
        MAY,
        JUNE,
        JULY,
        AUGUST,
        SEPTEMBER,
        OCTOBER,
        NOV,
        DEC,
        COMPLETED_TASK,
        TOTAL_TASK,
        DEPARTMENT,
        EMPLOYEES,
        KPI,
        CATEGORY_BREAKDOWN,
        MEASUREMENTS,
        ID,
        LOAD_DATE
    )
    VALUES (
        source.ID_TASKS,
        source.TYPE,
        source.TASKS,
        source.VALUE,
        source.JAN,
        source.FEB,
        source.MAR,
        source.APR,
        source.MAY,
        source.JUNE,
        source.JULY,
        source.AUGUST,
        source.SEPTEMBER,
        source.OCTOBER,
        source.NOV,
        source.DEC,
        source.COMPLETED_TASK,
        source.TOTAL,
        source.DEPARTMENT,
        source.EMPLOYEES,
        source.KPI,
        source.CATEGORY_BREAKDOWN,
        source.MEASUREMENTS,
        source.ID,
        source.LOAD_DATE
    );
```
